Project Characteristics Quadrants by Goal and Solution Uncertainty — Adapted from [1]

Type 1. Linear Strategy:

A linear strategy consist in sequential phases with no feedback loops. The project solution is not released until the final phase is reached. This strategy is characterized by clearly defined goal solution and requirements, zero or few change request of the scope, routine and repetitive process inside the project, use of pre-established formulas and templates.

Type 1. Incremental Strategy:

This strategy is similar to linear strategy but each phase of the project releases a partial solution. In this strategy the value must be deliver before the final phase.

Type 1 or 2. Iterative Strategy (e.g. Scrum or Feature-Driven Development)

Trying to fully map out and plan for an ever-changing environment is not efficient. Therefore, Agile encourages data-based, iterative decision making. For example, in Scrum, the primary focus is on delivering products that satisfy customer requirements in small iterative increments.

Similar to incremental strategy but consistent of a number of repeated phases. This strategy includes a feedback loop after a group of phases is completed. The loop stops when customer is satisfied, that is the strategy uses several intermediate solutions as a pathway to discover how the final solution should be.

Type 2 or 3. Adaptive Strategy (e.g. Adaptive Project Framework or Adaptive Software Development)

Similar to iterative strategy but each iteration feedback will adjust the future of project to converge over a complete solution. An iteration may create a partial solution to the customer. This strategy is adaptive because the final solution is only partially known by the customer, so the success depend on the ability to allocate frequent changes. Here planning is done in just-in-time manner. [2]

Type 4. Extreme Strategy (e.g. Research & Development projects)

Similar to adaptive strategy but each iteration feedback, the goal of the project must also be discover and the project will converge upon it. This is a type 4 strategy because nor is the goal or the solution process known. The level of uncertainty of goals and path in this kind of projects lead to high level of complexity. Usually final product is very different from what may be expected in the original intent.

```python
LIMIT=100

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius,
            LIMIT
            )
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Category']
    
    return(nearby_venues)
```
